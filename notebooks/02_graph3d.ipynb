# 3D Surface Animation

Animate a rotating 3D surface `f(x, y)`.

**Example equations:**
- `sin(sqrt(x**2 + y**2))` — ripple
- `exp(-0.1*(x**2 + y**2)) * cos(x + y)` — Gaussian-modulated wave
- `sin(x) * cos(y)` — saddle-wave
- `x * exp(-x**2 - y**2)` — Ricker wavelet

```python
import matplotlib
matplotlib.rcParams['animation.embed_limit'] = 50

import mathplt.animations
from mathplt.jupyter.widgets import EquationWidget, AnimationWidget
from mathplt.animations.graph3d import Graph3DAnimator
from mathplt.core.animator import AnimationConfig
```

```python
# Choose equation f(x, y)
eq = EquationWidget(variable='x', placeholder='sin(sqrt(x**2 + y**2))', label='f(x,y) =')
eq.display()
```

```python
config = AnimationConfig(fps=24, duration_seconds=10.0, figsize=(10, 7), dark_mode=True)

animator = Graph3DAnimator(
    config,
    equation=eq.equation,
    x_range=(-5.0, 5.0),
    y_range=(-5.0, 5.0),
    resolution=50,   # lower = faster render
    azim_per_frame=1.5,
)
AnimationWidget(animator).display()
```

```python
# Another preset: Gaussian-modulated wave
config2 = AnimationConfig(fps=24, duration_seconds=8.0, figsize=(10, 7))
anim2 = Graph3DAnimator(
    config2,
    equation='exp(-0.15*(x**2 + y**2)) * cos(x + y)',
    x_range=(-4, 4), y_range=(-4, 4), resolution=60,
    cmap='plasma',
)
AnimationWidget(anim2).display()
```
